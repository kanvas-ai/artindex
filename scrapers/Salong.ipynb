## Fetch art piece specific url sites, extract all useful info

```python
from urllib.request import urlopen, URLError
from bs4 import BeautifulSoup
from socket import timeout



base_url = "https://www.e-kunstisalong.ee/"
def get_soup(url: str):
    page = urlopen(url, timeout=1)
    html = page.read().decode("unicode_escape")
    soup = BeautifulSoup(html, "html.parser")
    return soup

def get_soup_persistent(url: str):
    haus_year_soup = None
    while True:
        try:
            haus_year_soup = get_soup(url)
            return haus_year_soup
        except timeout as e:
            print(e)
        except URLError as e:
            print(e)  
def get_salong_auction(url: str):
    art_pieces = []
    print(url)
    auc_soup = get_soup_persistent(url)
    auc_items = auc_soup.find_all("div", {"class":"tooteruut"})
    for item in auc_items:
        art_piece_obj = dict()
        item_info = item.find("div", {"class":"autor"}).a
        
        fields = item_info.find_all("span")
        author = fields[0].get_text()
        title = fields[1].get_text()
        desc = fields[2].get_text()
        prices = fields[3].get_text()
        
        url = item_info["href"]
        img = item.a.img["src"]
        art_piece_obj["author"] = author
        art_piece_obj["title"] = title
        art_piece_obj["desc"] = desc
        art_piece_obj["prices"] = prices
        art_piece_obj["url"] = url
        art_piece_obj["img"] = img
        art_pieces.append(art_piece_obj)
    return art_pieces

def get_salong_all():
    salong_art_pieces = []
    url_end = "E-kunstisalongi_oksjonid_756"
    print(base_url + url_end)
    allee_soup = get_soup_persistent(base_url + url_end)
    auctions = allee_soup.find_all("span", {"class":"alapealkiri"})
    auctions.pop() # last element is "Viimati lisatud teosed"
    for auc in auctions:
        auc_url = base_url + auc.a["href"]
        text = auc.a.get_text("|")
        texts = text.split("|")

        art_pieces = get_salong_auction(auc_url)
        for art_p in art_pieces:
            art_p["auction_name"] = texts[0]
            art_p["date"] = texts[1]
        salong_art_pieces.extend(art_pieces)
    return salong_art_pieces
            
data = get_salong_all()
    
    
```

## Clean df, remove unwanted characters

```python
import re
# Year from desc
df["year"] = df["desc"].str.extract("((19|20)\d{2})", expand=True)[0]
df["year_start"] = df["year"]
# sizes from desc
size  = df["desc"].str.extract(r"(\d+(\,\d+)?(\s)?[xX](\s)?\d+(\,\d+)?)", expand=True, flags=re.IGNORECASE)
#comma_number="\d+(\,\d+)?"
df["size"] = size[0].fillna(size[1]).fillna(size[2])
sizes = df["size"].str.split(r"[xX]", expand=True)

df["height"] = sizes[0].str.replace(",", ".")
df["width"] = sizes[1].str.replace(",", ".")

#df["tech"] = df["desc"].str.extract("([\s\.][A-ZÕÜÖÄ](\D)+?[\d$])", expand=True)[0]
# remove everything up to tech # remove everything after tech
df["tech"] = df["desc"].str.replace("(^(.)*((19|20)\d{2})(.)*?[\.\s])", "").str.replace("(\d(.)*x(.)*$)", "")
print(df['tech'])
df
#df["tech"] = df["desc"].str.split(".", 4, expand=True)[0]
df["date"] = df["date"].str.extract("((19|20)\d{2})", expand=True)[0]

# Prices
df["start_price"] = df["prices"].str.split(",", 1, expand=True)[0]
price_replaces = [r"(\((.)+\))", r"(\d*tk)", "€", "paar", " "]
start_prices = df["start_price"].str.split("[:]", 1, expand=True)[1]
for r in price_replaces:
    start_prices = start_prices.str.replace(r, "")
start_prices = start_prices.str.split("/", 1, expand=True)
print(start_prices)
start_prices[0] = pd.to_numeric(start_prices[0], errors="coerce", downcast="integer")
start_prices[1] = pd.to_numeric(start_prices[1], errors="coerce", downcast="integer")
print(start_prices)
df["start_price"] = start_prices.min(axis=1)

df["end_price"] = df["prices"].str.split(",", 1, expand=True)[1]
end_prices = df["end_price"].str.split(":", 1, expand=True)[1]
for r in price_replaces:
    end_prices = end_prices.str.replace(r, "")
end_prices = end_prices.str.split("/", 1, expand=True)
end_prices[0] = pd.to_numeric(end_prices[0], errors="coerce", downcast="integer")
end_prices[1] = pd.to_numeric(end_prices[1], errors="coerce", downcast="integer")
df["end_price"] = end_prices.min(axis=1)
print(df["end_price"].sum())
print(end_prices.sum())
print(end_prices[0][:1200].sum())
print(end_prices[1][1200:].sum())


# for col in ["start_price", "end_price"]:
#     df[col] = pd.to_numeric(df[col], errors="coerce")

#print(size)
#df.head()
df[["tech", "desc"]]
#df[df["start_price"].isnull()]
```
